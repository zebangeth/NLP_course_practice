```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
syn_data = pd.read_csv('syn_finetune_loss.csv')
real_data = pd.read_csv('real_finetune_loss.csv')
```

```python
syn_data
```

```python
# line plot of training loss
plt.figure(figsize=(10, 6))
plt.plot(syn_data['Step'], syn_data['Training_Loss'], label='Synthetic')
plt.plot(real_data['Step'], real_data['Training_Loss'], label='Real')
plt.title('Training Loss')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.legend()
```
